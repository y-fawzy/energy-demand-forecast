## 1. How did annual energy demand Change between 2015 and 2018? 

```python
annual=energy['total load actual']
annual.resample('Y').sum().plot.line(figsize=[20,10])
plt.xlabel("Year")
plt.ylabel("Energy Demand(MW)");
```

Energy demand has been steadily increasing between 2015 to 2018 but what is the rate of increase? It looks like the rate has slowed down between 2017 and 2018.

```python
yearly_dem=annual.resample('Y').sum()
perc_increase={"2015->16":(yearly_dem[1]-yearly_dem[0])*100/yearly_dem[0],
               "2016-->17":(yearly_dem[2]-yearly_dem[1])*100/yearly_dem[1],
              "2017-->18":(yearly_dem[3]-yearly_dem[2])*100/yearly_dem[2]}
perc_increase=pd.Series(perc_increase)
perc_increase.plot.line(figsize=[10,10])
plt.ylabel("% Increase")
plt.xlabel("Year to Year")
plt.title("Percentage change of energy demand");
```

Indeed the rate of increase has been going down from 2015 to 2018. The rate was around 1.05% in 2015 but has decreased to around 0.7%. Not a significant decreasebut signifies that the population is finding more efficient ways to utilize energy.

Let's dig deeper and look at monthly variations across 2015-->2018

## 2. How does energy demand vary by month?

```python
en_dem=energy['total load actual']
en_dem2015=en_dem[en_dem.index.year==2015].resample('M').sum()
en_dem2016=en_dem[en_dem.index.year==2016].resample('M').sum()
en_dem2017=en_dem[en_dem.index.year==2017].resample('M').sum()
en_dem2018=en_dem[en_dem.index.year==2018].resample('M').sum()
d_by_year=pd.DataFrame({2015:en_dem2015.values, 2016:en_dem2016.values, 2017:en_dem2017.values, 2018:en_dem2018.values})
d_by_year.index =np.arange(1,13)
```

```python
d_by_year.plot.line(figsize=[20,10])
plt.xlabel("Month")
plt.ylabel("Energy Demand(MW)")
plt.title('Total Monthly Demand by Year');
```

There are similar patterns across the 4 years and as expected energy demand peaks during winter(Heating Demand) and summer(Cooling demand). There is a sharp decrease between January and February but that is due to Febrary having less days(2-3) than January/March which is around 10% of the month. 

The mean of daily demand over each month would provide a better looking pattern.

```python
en_dem=energy['total load actual']
en_dem2015=en_dem[en_dem.index.year==2015].resample('M').mean()
en_dem2016=en_dem[en_dem.index.year==2016].resample('M').mean()
en_dem2017=en_dem[en_dem.index.year==2017].resample('M').mean()
en_dem2018=en_dem[en_dem.index.year==2018].resample('M').mean()
d_by_year=pd.DataFrame({2015:en_dem2015.values, 2016:en_dem2016.values, 2017:en_dem2017.values, 2018:en_dem2018.values})
d_by_year.index =np.arange(1,13)
d_by_year.plot.line(figsize=[20,10])
plt.xlabel("Month")
plt.ylabel("Energy Demand(W)")
plt.title('Average Daily Demand by Month for 2015-2018');
```

Much better!!

## Winter: 
A spike in demand occurs during winter(anytime in Jan).People need to warm up during the cold temperatures. Looks like winter has the highest demand.

## Spring/Autumn:
Demand starts decreasing during Feb-March-April-Early May where we have annual demand lows. Weather is usually mild and the population is starting to enjoy natural weather.

## Summer
Demand suddenly increase when the weather becomes too hot and humid to bear. We see another spike during June/July.

## Fall
Another low of the year occuring during September/October due to the transition into winter.

# During an average year, there will be 2 peaks and 2 lows for energy demand.


A plot of Hourly demand during 24 hours can better define the peak periods of daily demand.

## 3.  How does hourly demand change during the day during the 4 seasons?

```python
winter_dem=en_dem[(en_dem.index.month == 1) | (en_dem.index.month == 2)| (en_dem.index.month == 3)]
spring_dem=en_dem[(en_dem.index.month == 4) | (en_dem.index.month == 5)| (en_dem.index.month == 6)]
summer_dem=en_dem[(en_dem.index.month == 7) | (en_dem.index.month == 8)| (en_dem.index.month == 9)]
fall_dem=en_dem[(en_dem.index.month == 10) | (en_dem.index.month == 11)| (en_dem.index.month == 12)]
winter_hourly=winter_dem.groupby(winter_dem.index.hour).mean()
spring_hourly=spring_dem.groupby(spring_dem.index.hour).mean()
summer_hourly=summer_dem.groupby(summer_dem.index.hour).mean()
fall_hourly=fall_dem.groupby(fall_dem.index.hour).mean()
seasons=pd.DataFrame({"Winter":winter_hourly,
             "Spring":spring_hourly,
            "Summer":summer_hourly,
            "Fall":fall_hourly})
seasons.plot.line(figsize=[20,10])
plt.axvspan(19.5,20.5,alpha=0.3,color='r')
plt.axvspan(11,12,alpha=0.3,color='r')
plt.axvspan(2.5,4.5,alpha=0.3,color='g')
plt.axvspan(15.5,17.5,alpha=0.3,color='g')
plt.axhline(winter_hourly.max(),linestyle="--",color='b',linewidth=3)
plt.axhline(spring_hourly.max(),linestyle="--",color='y',linewidth=3)
plt.axhline(summer_hourly.max(),linestyle="--",color='g',linewidth=3)
plt.axhline(fall_hourly.max(),linestyle="--",color='r',linewidth=3)
plt.legend()
plt.xlabel('Hour of Day')
plt.ylabel('Energy Demand(W)')
plt.title("Hourly Energy Demand by Season")
plt.xticks(list(winter_hourly.index.values));
```

## Peaks and Lows of the Day:

The day ends up having 2 peaks of demand(Shaded in Red):
 - **First Peak**: Usually occurs between 11am - 12pm during most seasons. There is a broader distribution with winter but a sharper one with summer.  This demand peak occurs when everyone is at work and is right before lunch hour.
 - **Second Peak**: Usuall occurs between 7:30pm to 8:30pm and is consistent in all seasons.
 
 The day ends up having 2 low of demand(Shaded in Green):
 - **First Low**: Occurs between 2am-4:30am depending on the season. This probably when the majority of the population is sleeping.
 - **Second Low**: Occurs between 3:30pm and 4pm. This is when the majority of the population leave their offices/schools/commuting home/starting their naps. Summer has relatively high demand during this period. This could be due to the high temperatures or the extra sunlight that motivates people to more activities other than nap.
 
It can be expected that price peaks and lows will be similar to demand. Let's analyze...

```python
en_price=energy['price actual']
winter_price=en_price[(en_dem.index.month == 1) | (en_dem.index.month == 2)| (en_dem.index.month == 3)]
spring_price=en_price[(en_dem.index.month == 4) | (en_dem.index.month == 5)| (en_dem.index.month == 6)]
summer_price=en_price[(en_dem.index.month == 7) | (en_dem.index.month == 8)| (en_dem.index.month == 9)]
fall_price=en_price[(en_dem.index.month == 10) | (en_dem.index.month == 11)| (en_dem.index.month == 12)]
winter_p_hourly=winter_price.groupby(winter_dem.index.hour).mean()
spring_p_hourly=spring_price.groupby(spring_dem.index.hour).mean()
summer_p_hourly=summer_price.groupby(summer_dem.index.hour).mean()
fall_p_hourly=fall_price.groupby(fall_dem.index.hour).mean()
seasons_price=pd.DataFrame({"Winter":winter_p_hourly,
             "Spring":spring_p_hourly,
            "Summer":summer_p_hourly,
            "Fall":fall_p_hourly})
seasons_price.plot.line(figsize=[20,10])
plt.axvspan(19.5-1,20.5-1,alpha=0.3,color='r')
plt.axvspan(11-1,12-1,alpha=0.3,color='r')
plt.axvspan(2.5-1,4.5-1,alpha=0.3,color='g')
plt.axvspan(15.5-1,17.5-1,alpha=0.3,color='g')
plt.xticks(list(winter_hourly.index.values));
```

We see that the peaks and lows of prices occur 1-2 hours prior to peak demand hours.

```python
en_gen_ym=energy.drop(["time",'forecast solar day ahead',
       'forecast wind offshore eday ahead', 'forecast wind onshore day ahead',
       'total load forecast', 'total load actual', 'price day ahead',
       'price actual'], axis=1)
en_gen_ym_2015=en_gen_ym[en_gen_ym['year']==2015].sum().drop(['year','month'])
en_gen_ym_2016=en_gen_ym[en_gen_ym['year']==2016].sum().drop(['year','month'])
en_gen_ym_2017=en_gen_ym[en_gen_ym['year']==2017].sum().drop(['year','month'])
en_gen_ym_2018=en_gen_ym[en_gen_ym['year']==2018].sum().drop(['year','month'])
all_years_en_gen=[en_gen_ym_2015, en_gen_ym_2016, en_gen_ym_2017,en_gen_ym_2018]
en_gen_ym
```

```python
def remove_zero_generation(energy_gen_year):
    for idx in energy_gen_year.index:
        if energy_gen_year[idx]==0:
            energy_gen_year=energy_gen_year.drop(idx)
    return energy_gen_year
#OR-CHANGE-TO-NAN
```

```python
def normalize(energy_gen_year):
    all_energy=np.sum(energy_gen_year)
    source_perc={}
    for idx in energy_gen_year.index:
        source_perc[idx]=energy_gen_year[idx]*100/all_energy
    return pd.Series(source_perc)
```

```python
en_gen_ym_2015 = remove_zero_generation(en_gen_ym_2015)
en_gen_ym_2016 = remove_zero_generation(en_gen_ym_2016)
en_gen_ym_2017 = remove_zero_generation(en_gen_ym_2017)
en_gen_ym_2018 = remove_zero_generation(en_gen_ym_2018)
```

```python
en_gen_ym_2015=normalize(en_gen_ym_2015)
en_gen_ym_2016 = normalize(en_gen_ym_2016)
en_gen_ym_2017 = normalize(en_gen_ym_2017)
en_gen_ym_2018 = normalize(en_gen_ym_2018)
```

```python
x=en_gen_ym_2015.index
_X = np.arange(len(x))
plt.figure(figsize=(20,10))
plt.bar(_X-0.16, en_gen_ym_2015.values, 0.4, label='2015')
plt.bar(_X-0.06, en_gen_ym_2016.values, 0.4,label='2016')
plt.bar(_X+0.06, en_gen_ym_2017.values, 0.4,label='2017')
plt.bar(_X+0.16, en_gen_ym_2018.values, 0.4,label='2018')
plt.legend()
plt.xticks(_X, x, rotation=45)
plt.ylabel("% of Energy Source")
plt.show()
# MULTIPLE SUBPLOTS
```
